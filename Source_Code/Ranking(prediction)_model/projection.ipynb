```python
import torch

# Vectors
u1 = torch.tensor([1.0, 1.0])  # Vector u1
u2 = torch.tensor([1.0, -1.0])  # Vector u2
v = torch.tensor([1.0, 0.0])  # Unit vector along x-axis

# Dot product to compute projection
proj_u1 = torch.dot(u1, v)  # Projection of u1 onto x-axis
proj_u2 = torch.dot(u2, v)  # Projection of u2 onto x-axis

# Print results
print(f"Projection of u1 onto x-axis: {proj_u1.item()}")
print(f"Projection of u2 onto x-axis: {proj_u2.item()}")
```

```python
import torch
import matplotlib.pyplot as plt

# Vectors
u1 = torch.tensor([1.0, 1.0])  # Vector u1
u2 = torch.tensor([1.0, -1.0])  # Vector u2
v = torch.tensor([1.0, 0.0])  # Unit vector along x-axis

# Projections
proj_u1 = torch.dot(u1, v) * v  # Projection vector of u1 onto x-axis
proj_u2 = torch.dot(u2, v) * v  # Projection vector of u2 onto x-axis

# Plot setup
plt.figure(figsize=(8, 8))

# Plot the x-axis
plt.axhline(0, color='gray', linestyle='-', linewidth=1)
plt.axvline(0, color='gray', linestyle='-', linewidth=1)

# Plot the original vectors
plt.quiver(0, 0, u1[0], u1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='u1')
plt.quiver(0, 0, u2[0], u2[1], angles='xy', scale_units='xy', scale=1, color='green', label='u2')

# Plot the projections
plt.quiver(0, 0, proj_u1[0], proj_u1[1], angles='xy', scale_units='xy', scale=1, color='red', alpha=0.7, label='Projection of u1')
plt.quiver(0, 0, proj_u2[0], proj_u2[1], angles='xy', scale_units='xy', scale=1, color='black', alpha=0.7, label='Projection of u2')

# Customize plot
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.grid(alpha=0.3)
plt.title("Projection of Vectors onto the x-axis", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.legend()
plt.show()
```

```python
import torch
import matplotlib.pyplot as plt

# Vectors
u1 = torch.tensor([1.0, 1.0])  # Vector u1
u2 = torch.tensor([1.0, -1.0])  # Vector u2
v = torch.tensor([1.0, 0.0])  # Unit vector along x-axis

# Projections
proj_u1 = torch.dot(u1, v) * v  # Projection vector of u1 onto x-axis
proj_u2 = torch.dot(u2, v) * v  # Projection vector of u2 onto x-axis

# Figure 1: Projection of u1
plt.figure(figsize=(6, 6))
plt.axhline(0, color='gray', linestyle='-', linewidth=1)
plt.axvline(0, color='gray', linestyle='-', linewidth=1)
plt.quiver(0, 0, u1[0], u1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='u1')
plt.quiver(0, 0, proj_u1[0], proj_u1[1], angles='xy', scale_units='xy', scale=1, color='black', alpha=0.7, label='Projection of u1')
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.grid(alpha=0.3)
plt.title("Projection of u1 onto x-axis", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.legend()
plt.show()

# Figure 2: Projection of u2
plt.figure(figsize=(6, 6))
plt.axhline(0, color='gray', linestyle='-', linewidth=1)
plt.axvline(0, color='gray', linestyle='-', linewidth=1)
plt.quiver(0, 0, u2[0], u2[1], angles='xy', scale_units='xy', scale=1, color='green', label='u2')
plt.quiver(0, 0, proj_u2[0], proj_u2[1], angles='xy', scale_units='xy', scale=1, color='red', alpha=0.7, label='Projection of u2')
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.grid(alpha=0.3)
plt.title("Projection of u2 onto x-axis", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.legend()
plt.show()

# Figure 3: Both u1 and u2 with projections
plt.figure(figsize=(6, 6))
plt.axhline(0, color='gray', linestyle='-', linewidth=1)
plt.axvline(0, color='gray', linestyle='-', linewidth=1)
plt.quiver(0, 0, u1[0], u1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='u1')
plt.quiver(0, 0, u2[0], u2[1], angles='xy', scale_units='xy', scale=1, color='green', label='u2')
plt.quiver(0, 0, proj_u1[0], proj_u1[1], angles='xy', scale_units='xy', scale=1, color='black', alpha=0.7, label='Projection of u1')
plt.quiver(0, 0, proj_u2[0], proj_u2[1], angles='xy', scale_units='xy', scale=1, color='red', alpha=0.7, label='Projection of u2')
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.grid(alpha=0.3)
plt.title("Projection of u1 and u2 onto x-axis", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.legend()
plt.show()
```
